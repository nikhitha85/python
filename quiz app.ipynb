```python
class Quiz:
    def __init__(self, questions):
        self.questions = questions
        self.score = 0

    def display_question(self, question, index):
        print(f"Question {index + 1}: {question['text']}")
        for i, option in enumerate(question['options'], start=1):
            print(f"{i}. {option}")

    def get_user_answer(self):
        while True:
            try:
                answer = int(input("Enter your answer (1-4): "))
                if answer < 1 or answer > 4:
                    raise ValueError
                return answer
            except ValueError:
                print("Invalid input. Please enter a number between 1 and 4.")

    def check_answer(self, question, user_answer):
        correct_answer = question['correct']
        if user_answer == correct_answer:
            print("Correct!")
            self.score += 1
        else:
            print(f"Incorrect. The correct answer was {correct_answer}: {question['options'][correct_answer - 1]}.")

    def run(self):
        print("Welcome to the Quiz!\n")
        for index, question in enumerate(self.questions):
            self.display_question(question, index)
            user_answer = self.get_user_answer()
            self.check_answer(question, user_answer)
            print()  # Add a blank line for readability

        print(f"Quiz Complete! Your final score is {self.score}/{len(self.questions)}.")

# Customize the quiz questions here
questions = [
    {
        "text": "What is the capital of France?",
        "options": ["Berlin", "Madrid", "Paris", "Rome"],
        "correct": 3
    },
    {
        "text": "Which planet is known as the Red Planet?",
        "options": ["Earth", "Mars", "Jupiter", "Venus"],
        "correct": 2
    },
    {
        "text": "What is the largest mammal?",
        "options": ["Elephant", "Blue Whale", "Giraffe", "Polar Bear"],
        "correct": 2
    }
]

# Run the quiz
quiz = Quiz(questions)
quiz.run()
```
